```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.boxplot(data=df, y="math score");
```

```python
sns.boxplot(data=df, y="math score", x="test preparation course");
```

```python
plt.figure(figsize=(10,5),dpi=100)
sns.boxplot(data=df, y="math score", x="test preparation course");
```

```python
plt.figure(figsize=(10,5),dpi=100)
sns.boxplot(data=df, y="reading score", x="test preparation course");
```

```python
plt.figure(figsize=(10,5),dpi=100)
sns.boxplot(data=df, y="reading score", x="parental level of education");
```

```python
plt.figure(figsize=(10,5),dpi=100)
sns.boxplot(data=df, y="reading score", x="parental level of education", hue="test preparation course");
```

```python
plt.figure(figsize=(10,5),dpi=100)
sns.boxplot(data=df, y="reading score", x="parental level of education", hue="test preparation course")
plt.legend(bbox_to_anchor=(1.05,0.5));
```

```python
plt.figure(figsize=(10,8),dpi=100)
sns.boxplot(data=df, y="parental level of education", x="reading score", hue="test preparation course")
plt.legend(bbox_to_anchor=(1.05,0.5));
```

```python
plt.figure(figsize=(10,8),dpi=100)
sns.boxplot(data=df, y="parental level of education", x="reading score", hue="test preparation course",palette="Set2")
plt.legend(bbox_to_anchor=(1.05,0.5));
```

```python
plt.figure(figsize=(10,8),dpi=100)
sns.violinplot(data=df, y="parental level of education", x="reading score")
#plt.legend(bbox_to_anchor=(1.05,0.5));
```

```python
plt.figure(figsize=(10,8),dpi=100)
sns.violinplot(data=df, y="parental level of education", x="reading score",hue="test preparation course",palette="Set2")
plt.legend(bbox_to_anchor=(1.2,0.5));
```
