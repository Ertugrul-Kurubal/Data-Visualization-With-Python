```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_autos.csv")
df.head()
```

```python
df.info()
```

```python
df.groupby("vehicleType")[["vehicleType","price"]].min()
```

```python
vehicle_type_range = df.groupby("vehicleType")[["vehicleType","price"]].max()
vehicle_type_range
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x=vehicle_type_range.index,y=vehicle_type_range.price)
```

```python
vehicle_type_mean = df.groupby("vehicleType")[["vehicleType","price"]].mean()
vehicle_type_mean
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x=vehicle_type_mean.index,y=vehicle_type_mean.price)
```
