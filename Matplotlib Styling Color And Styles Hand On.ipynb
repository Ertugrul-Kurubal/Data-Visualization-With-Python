```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x);
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color="lightblue"); # RGB HEX Code
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color="#c7c47b")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color="#c7c47b")
ax.plot(x,x+1, color="purple")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color="#c7c47b", label="x vs x")
ax.plot(x,x+1, color="purple", label="x vs x+1")
ax.legend(); # ; equal to plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color="#c7c47b", label="x vs x", lw=5)
ax.plot(x,x+1, color="purple", label="x vs x+1", linewidth=0.5)
ax.legend(); 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color="#c7c47b", label="x vs x", lw=5, ls="--")
ax.plot(x,x+1, color="purple", label="x vs x+1", linewidth=2, linestyle="--")
ax.legend();
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color="#c7c47b", label="x vs x", lw=5, ls="--")
lines = ax.plot(x,x+1, color="purple", label="x vs x+1", linewidth=2, linestyle="--")
ax.legend();
```
