```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
labels = ['UK', 'USA', 'GER', 'JAP', 'FRA']
age_men = [40, 35, 45, 53, 28]
age_women = [35, 32, 48, 48, 31]
```

```python
fig, ax = plt.subplots(figsize=(6,4), dpi=200)
x = np.arange(len(labels))  
width = 0.4 
ax.bar(x,age_men,width,label="Men")
ax.set_ylabel('Ages')
ax.set_title('Ages means by gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc=1);
```

```python
fig, ax = plt.subplots(figsize=(6,4), dpi=200)
x = np.arange(len(labels))  
width = 0.4 
ax.bar(x,age_women,width,label="Women")
ax.set_ylabel('Ages')
ax.set_title('Ages means by gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc=1);
```
