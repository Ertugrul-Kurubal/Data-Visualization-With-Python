```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
fig.subplots_adjust()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
fig.subplots_adjust(wspace=0.9)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
fig.subplots_adjust(wspace=1, hspace=1)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_xlabel("X_LABEL 1,0")
axes[0][1].plot(x,y)
axes[0][1].set_xlim(2,6)
axes[1][0].plot(x,y)
axes[1][0].set_ylabel("Y_LABEL 1,0")
axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE')
fig.suptitle("Figure Level",fontsize=16)
plt.tight_layout()
fig.savefig('new_subplots.png',bbox_inches='tight')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(6,10),dpi=200)
axes[0][0].plot(x,y)
axes[0][0].set_xlabel("X_LABEL 1,0")
axes[0][1].plot(x,y)
axes[0][1].set_xlim(2,6)
axes[1][0].plot(x,y)
axes[1][0].set_ylabel("Y_LABEL 1,0")
axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE')
fig.suptitle("Figure Level",fontsize=16)
plt.tight_layout()
fig.set_figwidth(10)
fig.savefig('new_subplots.png',bbox_inches='tight')
```
